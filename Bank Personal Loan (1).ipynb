## **Column Descriptions:**


**ID:**
- Unique identifier for each customer. 

**Age:** 
- Age of the customer.

**Experience:**
- Years of professional experience.

**Income:**
- Annual income of the customer (in $000s).

**ZIP Code:**
- The ZIP code of the customer's residence.

**Family:**
- Number of family members in the household.

**CCAvg:**
- Average spending on credit cards per month (in $000s).

**Education:**
- Level of education (1: Undergraduate, 2: Graduate, 3: Advanced/Professional).

**Mortgage:**
- Value of house mortgage if any (in $000s).

**Personal Loan:**
- Whether the customer accepted a personal loan offer (1 = Yes, 0 = No).

**Securities Account:**
- Does the customer have a securities account? (1 = Yes, 0 = No).

**CD Account:**
- Does the customer have a certificate of deposit (CD) account? (1 = Yes, 0 = No).

**Online:**
- Does the customer use internet banking? (1 = Yes, 0 = No).

**CreditCard:**
- Does the customer have a credit card issued by the bank? (1 = Yes, 0 = No).
